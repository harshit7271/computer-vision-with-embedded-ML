```python
import numpy as np
import matplotlib.pyplot as plt
import PIL
```

```python
image_path = "/content/grayEle.bmp"
```

```python
# Open image as (grayscale) and convert it to a bumpyt array
img = PIL.Image.open(image_path).convert('L')            # L : 8 bit grayscale mode, converts any input image -RGB to a single channel image representing shades from black (0) to white (255)
print(img)
```

```python
# converting to numpy array
img_array = np.asarray(img)
print(img_array.shape)
print("(num rows, num columns)")
print()
print(img_array)
```

```python
# show normalized array (values between 0 and 1)
print(img_array / 255)
```

```python
# Display image from array
plt.imshow(img_array, cmap='gray', vmin=0, vmax=255)
```

```python
img_vector = img_array.reshape(1,20)
print(img_vector)
```
