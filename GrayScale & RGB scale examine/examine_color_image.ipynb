```python
import PIL
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = "/content/Turtle.bmp"
```

```python
# Open image as color scale and convert it to numpy array
img = PIL.Image.open(img_path).convert('RGB')
print(img)
```

```python
img_array = np.asarray(img)
print(img_array.shape)
print(img_array)
```

```python
## Display image from array

# Create 3 copies of the array
img_red = img_array.copy()
img_green = img_array.copy()
img_blue = img_array.copy()

# fill non red channels with 0s
img_red[:, :, 1].fill(0)     # Give everything from green(1st index col in matrix) channel and fill with 0
img_red[:, :, 2].fill(0)      # Give everything from blue(2nd index col in matrix) channel and fill with 0

# fill non green channels with 0s
img_green[:, :, 0].fill(0)
img_green[:, :, 2].fill(0)

# fill non blue channels with 0s
img_blue[:, :, 0].fill(0)
img_blue[:, :, 1].fill(0)

# PLot each of the RGB channels separately
plt.figure()
plt.imshow(img_red, vmin=0, vmax =255)
plt.figure()
plt.imshow(img_green,  vmin=0, vmax =255)
plt.figure()
plt.imshow(img_blue,  vmin=0, vmax =255)
plt.figure()
```

```python
# print all of these together
plt.figure()
plt.imshow(img_array, vmin=0, vmax =255)
plt.figure()
```
